```python
df.columns=(
            df.columns
            .str.strip()
            .str.lower()
            .str.replace(" ","_")
           )
```

```python
df = df.dropna(subset=["Order ID", "Amount"])
```

```python
df["Date"] = pd.to_datetime(
    df["Date"],
    format="%m-%d-%y",
    errors="coerce"
)

```

```python
df.columns
```

```python
df["Date"] = pd.to_datetime(df["Date"])

df["order_month"] = df["Date"].dt.month
df["order_year"] = df["Date"].dt.year
```
